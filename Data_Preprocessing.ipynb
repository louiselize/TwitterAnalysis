# II. Data Pre processing

```python
import pandas as pd
```

```python
filename = 'ukraine-query-tweets'
```

```python
# Lire le fichier CSV et créer un dataframe
tweets_df = pd.read_csv(filename +'.csv')
display(tweets_df)
```

## 1. Convert user json column into new columns

```python
import ast 

tweets_df.dropna(subset=['user'],inplace=True)
tweets_df['user'] = tweets_df['user'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
for key in tweets_df['user'][0].keys():
    tweets_df['user-'+key] = tweets_df['user'].apply(lambda x: x[key])
tweets_df
```

```python
tweets_df.info()
```

## 2. Row treatment

Remove tweets with less than 1 retweet

```python
mask = (tweets_df.retweetCount > 0)
tweets_df = tweets_df.loc[mask]
```

Remove tweets with link

```python
tweets_df.links = tweets_df.links.fillna(0) #remplacer na par 0
mask = (tweets_df.links == 0)
tweets_df = tweets_df.loc[mask]

tweets_df
```

Check duplications

```python
tweets_df = tweets_df.drop_duplicates(subset='rawContent')
tweets_df
```

## 3. New features

Tweet char count

```python
tweets_df['charCount'] = tweets_df['rawContent'].apply(lambda x: len(x))
tweets_df
```

Tweet emoji count

```python
import re
```

```python
# Function to count emojis in a string
def count_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    emojis = re.findall(emoji_pattern, text)
    return len(emojis)
```

```python

# Apply the emoji count function to each tweet
tweets_df['emojiCount'] = tweets_df['rawContent'].apply(count_emojis)

# Access the emoji count
tweets_df
```

Tweet Words Count

```python
# Function to count words in a string
def count_words(text):
    words = text.split()
    return len(words)
```

```python
tweets_df['wordCount'] = tweets_df['rawContent'].apply(count_words)
```

Tweet Emojis to char ratio

```python
def calculate_emoji_ratio(row):
    char_count = row['charCount']
    emoji_count = row['emojiCount']
    if char_count > 0:
        ratio = emoji_count / char_count
    else:
        ratio = 0
    return ratio
```

```python
tweets_df['emojiCharRatio'] = tweets_df.apply(calculate_emoji_ratio, axis=1)
```

Tweet word to char ratio

```python
# Function to calculate the ratio of words to characters
def calculate_word_ratio(row):
    word_count = row['wordCount']
    char_count = row['charCount']
    if char_count > 0:
        ratio = word_count / char_count
    else:
        ratio = 0
    return ratio
```

```python
# Apply the word ratio function to each row of the DataFrame
tweets_df['wordRatio'] = tweets_df.apply(calculate_word_ratio, axis=1)
```

Hastags count

```python
# Function to count hashtags in a string
def count_hashtags(text):
    hashtags = re.findall(r"#(\w+)", text)
    return len(hashtags)
```

```python
# Apply the hashtag count function to each tweet
tweets_df['hashtagCount'] = tweets_df['rawContent'].apply(count_hashtags)
```

```python
tweets_df[['charCount', 'emojiCount', 'emojiCharRatio', 'wordCount', 'wordRatio', 'hashtagCount']]
```

Author total activity

```python
tweets_df['user-totalActivity'] = (tweets_df['user-statusesCount'] + tweets_df['user-favouritesCount']) 
```

Author tweets per year

```python
#Convert the "date" column to a pandas datetime format:
tweets_df['date'] = pd.to_datetime(tweets_df['date'])
```

```python
import pandas as pd
from datetime import datetime, timezone

# Convert the 'user-created' column to datetime format
tweets_df['user-created'] = pd.to_datetime(tweets_df['user-created'])

# Convert the timezone of 'user-created' column to the desired timezone
tweets_df['user-created'] = tweets_df['user-created'].dt.tz_convert(timezone.utc)

# Calculate the account age in years
current_datetime = datetime.now(timezone.utc)
tweets_df['account_age'] = (current_datetime - tweets_df['user-created']).dt.days / 365

# View the updated DataFrame
print(tweets_df)
```

```python
# Create the new feature 'Author Tweets per Year'
tweets_df['user-tweetsPerYear'] = tweets_df['user-statusesCount'] / tweets_df['account_age']
```

Author Total Activity per year

```python
tweets_df['user-totalActivityPerYear'] = tweets_df['user-totalActivity'] / tweets_df['account_age']
```

# FEATURES FOR TEXTUAL ANALYSIS

```python
columns_to_add = ['retweetCount','rawContent']
```

```python
tweet_content_rt_predict_df = tweets_df[columns_to_add]
```

```python
tweet_content_rt_predict_df
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Télécharger les ressources nécessaires pour NLTK (exécuter une fois)
nltk.download('stopwords')
nltk.download('wordnet')

# Initialiser le lemmeur
lemmatizer = WordNetLemmatizer()

# Liste des mots vides
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Supprimer les caractères spéciaux
    text = re.sub(r'[^A-Za-z]', ' ', text)

    # Convertir en minuscules
    text = text.lower()

    # Diviser le texte en mots individuels
    words = text.split()

    # Supprimer les mots vides
    words = [word for word in words if word not in stop_words]

    # Lemmatisation des mots
    words = [lemmatizer.lemmatize(word) for word in words]

    # Rejoindre les mots en une seule chaîne de texte
    processed_text = ' '.join(words)

    return processed_text

# Appliquer la fonction de prétraitement à la colonne de tweets
tweets_df['processedText'] = tweets_df['rawContent'].apply(preprocess_text)
```

```python
tweets_df.processedText
```

```python
from collections import Counter

# Concaténer tous les textes prétraités en une seule chaîne de texte
all_text = ' '.join(tweets_df['processedText'])

# Diviser la chaîne de texte en mots individuels
words = all_text.split()

# Compter le nombre d'occurrences de chaque mot
word_counts = Counter(words)

# Triez les mots en fonction de leur fréquence d'occurrence
most_common_words = word_counts.most_common(20)  # Vous pouvez ajuster le nombre de mots affichés ici

# Afficher les mots les plus utilisés
for word, count in most_common_words:
    print(f'{word}: {count}')
```
